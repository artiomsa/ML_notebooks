```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
import pandas as pd
```

```python
clf = RandomForestClassifier(random_state = 0)
```

```python
par = {'n_estimators' : range(10,51,10) , 'max_depth' : range(1,13,2) , 'min_samples_leaf' : range(1,8)\
       , 'min_samples_split' : range(2,10,2)}
```

```python
searcher = GridSearchCV(clf,par,cv = 3,n_jobs=-1)
```

```python
test = pd.read_csv('/Users/artemarhipov/Downloads/testing_mush.csv')
data = pd.read_csv('/Users/artemarhipov/Downloads/training_mush.csv')
```

```python
data.head()
```

```python
searcher.fit(data.drop('class',axis = 1), data['class'])
```

```python
searcher.best_params_
```

```python
clf = searcher.best_estimator_
```

```python
xes = data.columns
```
